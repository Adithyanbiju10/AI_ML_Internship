```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/preprocessing_practice_dataset.csv")
```

```python
df.describe()
```

```python
df['Age_mean_imputed']=df['Age'].fillna(df['Age'].mean())
df['Age_median_imputed']=df['Age'].fillna(df['Age'].median())
df[['Age','Age_mean_imputed','Age_median_imputed']]
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Experience_Years',y='Monthly_Salary',palette='pastel')
plt.title('Boxplot of Experience Years vs Monthly Salary')
plt.xlabel('Experience Years')
plt.ylabel('Monthly Salary')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(21,6))
sns.boxplot(data=df,x='Age',y='Experience_Years',palette='pastel')
plt.title('Boxplot of Experience and Age')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(21,6))
sns.boxplot(data=df,x='Education',y='Monthly_Salary',palette='pastel')
plt.title('Boxplot of Education and Monthly Salary')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(21,6))
sns.boxplot(data=df,x='Department',y='Performance_Rating',palette='pastel')
plt.title('Boxplot of Department and Performance')
plt.show()
```
